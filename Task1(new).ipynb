```python
#Step 01: Importing the pandas Module
import pandas as pd

#Step 02: Importing the netflix dataset into pandas dataframe object
df = pd.read_csv("netflix_titles.csv")
```

```python
#Step 03: Extracting the First 5 rows of the dataset 
df.head()
```

```python
#Step 04: Gives summary of your DataFrame (non-null column values, data types, total memory usage)
df.info()
```

```python
#Step 05: Gives detailed statistical summary of DataFrame (Count of non-null column values, Unique values, Top value and its frequency, Mean, min, max, and standard deviation)
df.describe(include='all')
```

```python
#Step 06: Gives list of column names in DataFrame
df.columns
```

```python
#Step 07: Checking the overall no. of rows & columns in dataset
df.shape
```

```python
#Step 08: Checking for missing/null values
df.isnull().sum()
```

```python
#Step 09: For text columns like director, cast, filling with "Unknown"
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
```

```python
#Step 10: Rechecking
df.isnull().sum()
```

```python
#Step 11: Dropping NaN values of date_added cloumn because very few values
df.dropna(subset=['date_added'], inplace=True)
```

```python
#Step 12: Rechecking
df.isnull().sum()
```

```python
#Step 13: Replacing the country as unknown
df['country'].fillna('Unknown', inplace=True)
```

```python
#Step 12: Rechecking
df.isnull().sum()
```

```python
#Step 11: Dropping NaN values of rating & duration cloumn because very few values
df.dropna(subset=['rating','duration'], inplace=True)
```

```python
#Step 12: Rechecking
df.isnull().sum()
```

# ----- All values are replaced/filled/dropped -----


```python
#Step 13: Checking for duplicate Values
df.duplicated().sum()
```

```python
#Step 14: If there are any use the below
#df.drop_duplicates(inplace=True)
```

```python
#Step15 : Standardizing the text columns data
df['type'] = df['type'].str.strip().str.lower()
df['country'] = df['country'].str.strip().str.lower()
df['rating'] = df['rating'].str.strip().str.upper()
```

```python
#Step 15: Rechecking
df.head(10)
```

# ---Working with dates -----


```python
df.dtypes
```

```python
#Step 16:Changing the date_added cloumn to datetime object
df['date_added'] = pd.to_datetime(df['date_added'], dayfirst=True, errors='coerce')
```

```python
df.dtypes
```

```python
#Step 17: Wanting a certain format
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%y')
```

```python
#Step 15: Rechecking
df.head(10)
```

```python
#Step 18: Checking for any invalid dates
df['date_added'].isna().sum()
```

```python
#Step 19: Renaming columns to all lower and no spaces between them
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
#Step 20: Rechecking
df.head(10)
```

```python
#Step 21: The different/unique categories
df['rating'].unique()
```

```python
#Step 22: The frequency of the different/categories
df['rating'].value_counts()
```

```python
#Step 23: Converting to type category -> for less memory usage and faster operations on category data
df['rating'] = df['rating'].astype('category')
```

```python
#Step 23: Exporting the finally cleaned dataset 
df.to_csv("netflix_titles_cleaned.csv", index=False)
```
